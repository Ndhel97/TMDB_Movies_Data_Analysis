```python
import json
import pandas as pd
  
# Opening JSON file
f = open('tmdb_links.json', 'r')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
df = pd.DataFrame.from_dict(pd.json_normalize(data), orient='columns')
```

```python
df_new = df[['link', 'title', 'poster']]
df_new
```

```python
# Save dataframe into json file
df_new.to_json('tmdb_movies_data_new.json', orient='records', compression = 'infer')
```
