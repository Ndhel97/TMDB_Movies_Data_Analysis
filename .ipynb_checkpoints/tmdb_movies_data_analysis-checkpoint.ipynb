```python
import json
  
# Opening JSON file
f = open('tmdb_links.json',)
  
# returns JSON object as 
# a dictionary
data = json.load(f)
df = pd.DataFrame.from_dict(pd.json_normalize(data), orient='columns')
```

```python
df
```

```python
list(df.columns.values)
```

```python
for col in df.columns[11:]:
    if ('Writer' or 'Director') in col:
        continue
    elif 'person' not in col:
        del df[col]
df
```
